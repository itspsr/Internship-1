```python
## Connect to the driver : -
driver = webdriver.Chrome(r"C:/Users/swati/Downloads/chromedriver_win32/chromedriver.exe")
```

```python
## Finding element for job search
search_jobs =driver.find_element_by_id("qsb-keyword-sugg")
search_jobs.send_keys("Data Analyst")
search_loc = driver.find_element_by_id("qsb-location-sugg")
search_loc.send_keys("Bangalore")
```

```python
## Finding element for Search button
search_button = driver.find_element_by_xpath("//div[@class = 'search-btn']/button")
search_button.click()
```

```python
## Now Creating empty lists for Job -Title,Location,Company Name,Experience Required:
Job_Title = []
location =[]
company_name =[]
exp =[]
```

```python
## Scrapping Job Title :
job_title_tag = driver.find_elements_by_xpath("//a[@class ='title fw500 ellipsis']")

#job_title_tag[:10]

for i in job_title_tag:
    Job_Title.append(i.text)
print("****** Job Title****************")    
#print(Job_Title)    
Job_Title =Job_Title[0:2] +Job_Title[20:28]
print(Job_Title)
print(len(Job_Title))
print("*********************************")

#### Scrapping Location
#search_location = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span[1]")
search_location = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']")
#search_location[:10]
for i in search_location:
    location.append(i.text)
print("********** Location*************")  
#print(location)
location = location[0:2]+location[20:28]
print(location)
print(len(location))
print("********************************")    

### Scrapping Company Name :-
company = driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
for i in company:
    company_name.append(i.text)
    
print("********** Company Name *************")        
company_name=company_name[0:2]+company_name[20:28]
print(len(company_name))
print(company_name)
print(len(company_name))

print("********************************")    

## Scrapping Experience Required : 
exp_req = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']/span[1]")
for i in exp_req:
    exp.append(i.text)
    
print("********** Experience Required *************")        
exp = exp[0:2]+exp[20:28]    
print(exp)
print(len(exp))
print("********************************")    


data_analyst_df = pd.DataFrame({})
data_analyst_df['Job Title'] = Job_Title
data_analyst_df['Location'] = location
data_analyst_df['Company Name'] = company_name
data_analyst_df['Experience Required'] = exp


data_analyst_df
```

```python
## making empty lists for Company name, Total salary record,min salary,average salary,max salary and experience required:-

name =[]
total_salary=[]
min_salary =[]
max_salary =[]
avg_salary =[]
exp =[]

## Scrapping Company Name:
company = driver.find_elements_by_xpath("//div[@class='name']/a")
for i in company:
    name.append(i.text)
print("------------Company Name-------------------------")    
print(name)
print("------------------------------------------------")
## Since list has returned elements with single inverted comma,so using translate method removing it.
## For this ascii code of "'" i.e;39 is set to None.
target = {39:None}
print(str(name).translate(target))   
## Scrapping Total salary record:
t_salary = driver.find_elements_by_xpath("//div[@class='name']/span")
for i in t_salary:
    total_salary.append(i.text)
print("--------------Total Salary Record---------------------------")    
print(total_salary)   
###Removing small braces from Total Salary list using translate method.Since it is a string method so type casted total_salary into 
## string and then use ascii code for "'" that is 39, 40 for "(" and 41 for ") " and set it to none and then translate.

target ={39 : None,40 : None, 41 : None}
total_sal_rec =str(total_salary).translate(target)
print(total_sal_rec)

print("---------------------------------------------------------------")


### Scrappin min salary:-
#salary = driver.find_elements_by_id("//div[@class='result-col salary-range'")
m_salary = driver.find_elements_by_xpath("//div[@class='result-col salary-range']//div[@class='salary-values']//div[1]")

for i in m_salary:
    min_salary.append(i.text)
print("------------Minimum Salary----------------")    
print(min_salary)

## Since list has returned elements with single inverted comma,so using translate method removing it.
## For this ascii code of "'" i.e;39 is set to None.
print("------After removing single inverted comma-----------------")
target = {39:None}
print(str(min_salary).translate(target))   

ma_salary = driver.find_elements_by_xpath("//div[@class='result-col salary-range']//div[@class='salary-values']//div[2]")
for i in ma_salary:
    max_salary.append(i.text)
print(max_salary)   
## Since list has returned elements with single inverted comma,so using translate method removing it.
## For this ascii code of "'" i.e;39 is set to None.
print("------After removing single inverted comma-----------------")

target = {39:None}
print(str(max_salary).translate(target))   
print("-----------------------------")

av_salary = driver.find_elements_by_xpath("//div[@class='average-indicator-wrapper']//p[@class='averageCtc']")
for i in av_salary:
    avg_salary.append(i.text)
print(avg_salary[9:])   

## Since list has returned elements with single inverted comma,so using translate method removing it.
## For this ascii code of "'" i.e;39 is set to None.

print("------After removing single inverted comma-----------------")

target = {39:None}

print(str(avg_salary[9:]).translate(target))   
print("-----------------------------")

exp_req = driver.find_elements_by_xpath("//div[@class='salaries one-line sbold-list-header']/span")
for i in exp_req:
    exp.append(i.text)
print(exp) 

## Since list has returned elements with single inverted comma,so using translate method removing it.
## For this ascii code of "'" i.e;39 is set to None.

print("------After removing single inverted comma-----------------")

target = {39:None}
print(str(exp).translate(target))   
print("-----------------------------")

### Finally making dataframe:-
salary_df = pd.DataFrame({'Company Name':name,
                          'Total salary record':total_salary,
                          'Min Salary':min_salary,
                          'Max Salary':max_salary,
                          'Avg Salary':avg_salary[9:],
                           'Experience':exp
                          
    
                         })
salary_df
```
