```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
def find_house(link):
    
    
    page = requests.get(link)
    
    soup = BeautifulSoup(page.content)
    house_title_list =[]
    house_title = soup.find_all( 'h2',class_='heading-6 font-semi-bold nb__25Cl7')
    for i in house_title:
        house_title_list.append(i.text)
    print("***********House_title_list****************")    
    print(house_title_list)   
    print(len(house_title_list))
    print("*******************************************")
    
    location_list = []
    location = soup.find_all('div',class_='nb__1EwQz')
    for i in location:
        location_list.append(i.text)
    print("***********location_list****************")    
    
    print(location_list)
    print(len(location_list))
    print('****************************************')
    area_list =[]
    area = soup.find_all('div',class_='nb__FfHqA')
    for i in area:
        area_list.append(i.text)
    print("***********Area_list*********************")    
    
    print(area_list)    
    print('******************************************')    
    emi_list =[]
    emi = soup.find_all('div',class_='font-semi-bold heading-6')
    for i in emi:
        emi_list.append(i.text)
    print("************emi_list**********************")    
    print((emi_list[1::3]))
    print("******************************************") 
    price_list = []
    price = soup.find_all('div',class_='font-semi-bold heading-6')
    for i in price:
        price_list.append(i.text)
    print("********price_list****************")    
    print(price_list[2::3])    
    print("************************************")
    
    no_broker_df = pd.DataFrame({'House_title':house_title_list,
                                 'Location' :location_list,
                                  'Area':area_list,
                                   'Emi':emi_list[1::3],
                                    'Price':price_list[2::3]
                                 })
    print(no_broker_df)   

find_house("https://www.nobroker.in/property/sale/mumbai/Andheri%20East?searchParam=W3sibGF0IjoxOS4xMTc4NTQ4LCJsb24iOjcyLjg2MzEzMDQsInBsYWNlSWQiOiJDaElKTWJIZlFSdTI1enNSTWF6ZFkzVXBhS1kiLCJwbGFjZU5hbWUiOiJBbmRoZXJpIEVhc3QifV0=&radius=2.0&type=BHK2&city=mumbai&locality=Andheri%20East")
```
