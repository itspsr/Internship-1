```python
##Connect to the driver
driver = webdriver.Chrome(r"C:/Users/swati/Downloads/chromedriver_win32/chromedriver.exe")
time.sleep(3)

## Connect to the url
url = "https://github.com/"
driver.get(url)

driver.maximize_window()

time.sleep(2)

```

```python
## Scrapping Explore:
explore = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary")

#print(explore.get_attribute('href'))
try:
    explore.click()
except ElementNotInteractableException:
    driver.get(explore.get_attribute('href'))

    
```

```python

# ## scrapping Trending:

trending = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a")
try:
    trending.click()
except ElementNotInteractableException:
    driver.get(trending.get_attribute('href'))
```

```python
## Scrapping Details:-
##Scrapping Repository Title:-
names=[]
try:
    name = driver.find_elements_by_xpath("//*[@id='js-pjax-container']/div[3]/div/div[2]/article/h1/a")
    for i in name:
        names.append(i.text)
except NoSuchElementException:
    names.append('NA')
names = names[:22]
print(len(names))


## Scrapping Description
Description=[]
try:
    desc = driver.find_elements_by_xpath("//*[@id='js-pjax-container']/div[3]/div/div[2]/article/p")
    for i in desc:
        Description.append(i.text)
except NoSuchElementException:
    Description.append('NA')
    
Description = Description[:22]

## Scrapping language

Language=[]
try:
    lang = driver.find_elements_by_xpath("//*[@id='js-pjax-container']/div[3]/div/div[2]/article/div/span[1]/span[2]")
    for i in lang:
        Language.append(i.text)
except NoSuchElementException:
    Language.append('-')

print(len(Language))
    
## Scrapping urls to get contributor's count:    
urls=[]
match_page=driver.find_elements_by_xpath('//*[@id="js-pjax-container"]/div[3]/div/div[2]/article/h1/a')
for i in match_page:
    urls.append(i.get_attribute('href'))
    
print(len(urls))
    
    
counts=[]
try:
    for i in urls:
        driver.get(i)
        count =driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]')    
        list=count.text.split('\n')
        for j in list:
            if j.split()[0]=='Contributors':
                counts.append(j.split()[1])
                break
        else:
            counts.append('N/A')
except NoSuchElementException:
    counts.append('N/A')
counts = counts[:22]  

print(len(counts))




df = pd.DataFrame({"Repository Title":names[:22],
                   "Repository description":Description[:22],
                   "Contributors count":counts[:22],
                  "Language used":Language[:22]})
df
```
